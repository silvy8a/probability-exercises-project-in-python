## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# TODO
import random
import numpy as np

def roll_dice (iterations=1000):
    count=0
    for i in range (1000):
        dice_1 = random.randint(1,6)
        dice_2 = random.randint(1,6)
        total= dice_1 + dice_2 
        if (total >7 or total%2==0):
                count +=1
    return count / iterations

print(f'The probability of getting an even number or one greater than 7 is {round(roll_dice() *100,2)}%')

```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
import numpy as np
import random

def caja_bolas (iterations = 1000):
    combination = 0
    same_color = 0
    for _ in range(iterations):
        ball_color=[]
        
        for _ in range (5):
            ball_color.append(ball_box[np.random.randint(0, 59)])
        
        white_balls = ball_color.count("White")
        red_balls = ball_color.count("Red")
        green_balls = ball_color.count("Green")
                
        if (white_balls == 3) and (red_balls == 2):
            combination += 1
        if (white_balls == 5) or (red_balls == 5) or (green_balls == 5):
            same_color += 1

    return combination / iterations, same_color / iterations

results=caja_bolas(10000)

print(f"The probability of 3 white balls and 2 red ones is: {np.round(results[0] * 100, 2)}%")
print(f"The probability of all balls the same color is: {np.round(results[1] * 100, 2)}%")

```
